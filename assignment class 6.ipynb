### Task9:
Write a program that accepts a sentence and calculate the number of letters and digits.

Suppose the following input is supplied to the program:

hello world! 123

Then, the output should be:

LETTERS 10

DIGITS 3

```python
acount=0
lcount=0
import string
code=input('enter your sentence and digits\n')
for i in code:
    if i.isdigit():
        lcount+=1
    elif i.isalpha():
        acount+=1
    elif i==' ':
        continue
    elif i in string.punctuation:
        continue
print(f'the alphabet in the sentence is {acount} and letter is {lcount}')
```

### Task11: 
You need to write a program that counts the number of vowels in a given sentence as input from console. 

Suppose the following input is supplied to the program:

Hello world! Practice makes perfect

Then, the output should be:

a appeared 2 times

e appeared 5 times

i appeared  1 time

o appeared 2 times

u appeared 0 time

```python
sent=input('enter your sentence')
vowel='aeiou'
for i in vowel:
    if i=='a':
        print(f'the number (a) vowel is {sent.count(i)}')
    elif i=='e':
        print(f'the number of (e) vowel is {sent.count(i)}')
    elif i=='i':
        print(f'the number of (i) vowel is {sent.count(i)}')
    elif i=='o':
        print(f'the number of (o) vowel is {sent.count(i)}')
    elif i=='u':
        print(f'the number of (u) vowel is {sent.count(i)}')
```

```python
text=input('enter your text#n')
vowels='aeiou'
a=e=i=o=u=0
for vowel in vowels:
    for sent in text:
        if vowel==sent:
            if vowel=='a':
                a+=1
            elif vowel=='e':
                e+=1
            elif vowel=='i':
                i+=1
            elif vowel=='o':
                o+=1
            elif vowel=='u':
                u+=1
print(f'''  the number of a is {a}
            the number of e is {e}
            the number of i is {i}
            the number of o is {o}
            the number of u is {u}''')
```

### Task14:
You need code a program that calculates and prints the value according to the given formula:
P = Square root of [(2 * A * B)/C]

Following are the fixed values of A and B:

A is 50. B is 30.

The values of the literal C should be taken as console input to your program in a comma-separated sequence.

Example:
Let us assume the following comma separated input sequence is given to the program:
100,150,180

The output of the program should be:
18,22,24

```python
import math
A=50
B=30
p_val=[]
C=input('enter your number').split(',')
for c in C:
    P=(2*A*B)/int(c)**0.5
    p_val.append(math.floor(P))
print(p_val)
```
