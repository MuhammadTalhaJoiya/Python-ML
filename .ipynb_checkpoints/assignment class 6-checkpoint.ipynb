### Task4:
Write a program to display sum of odd numbers and even numbers that fall between 12 and 37(including both numbers)

```python
counte=0
counto=0
sume=0
sumo=0
for i in range(12,38):
    if i%2==0:
        counte+=1
        print(f'even number is {i}')
        sume+=i
        avge=sume/counte
    else:
        print(f'odd number {i}')
        counto+=1
        sumo+=i
        avgo=sumo/counto
print(f'sum of even number is {sume}')
print('the avg of even number is',avge)
print(f'sum of odd number is {sumo}')
print('the avg of odd number is',avgo)
```

### Task5:
Write a program to print table of a number accepted from user.

```python
table_user=int(input('enter your table number'))
for j in range(1,11):
    print(f'{table_user}*{j}={table_user*j}')
```

```python
table_user=input('enter your table number')
for i in table_user:
    for j in range(1,11):
        print(f'{i}*{j}={int(i)*j}')
```

### Task6:
Write a program to display product of the digits of a number accepted from the user.

If user inputs:123 Outputs 6

If user inputs: 567 Outputs 210

```python
product=1
digi=input('enter your digits number')
for i in digi:
    print(i)
    product*=int(i)
print(f'the product of digits is given by the user {product}')
```

### Task9:
Write a program that accepts a sentence and calculate the number of letters and digits.

Suppose the following input is supplied to the program:

hello world! 123

Then, the output should be:

LETTERS 10

DIGITS 3

```python
acount=0
lcount=0
import string
code=input('enter your sentence and digits\n')
for i in code:
    if i.isdigit():
        lcount+=1
    elif i.isalpha():
        acount+=1
    elif i==' ':
        continue
    elif i in string.punctuation:
        continue
print(f'the alphabet in the sentence is {acount} and letter is {lcount}')
```

### Task10: 
You need to trace and make a list of all such numbers from1000 to 3000 in which all the digits are even numbers.
